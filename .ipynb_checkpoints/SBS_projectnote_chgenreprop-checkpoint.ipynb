```python
import sys, os
import pandas as pd
import numpy as np
from datetime import timedelta
import datetime
from ggplot import *
```

```python
tidy_fin = pd.read_csv("tidy_data_with_demoinfo.csv")
```

```python
tidy_fin = tidy_fin.drop('Unnamed: 0', axis=1)
```

```python
tidy_fin
```

# 1. 최소시청시간 10분 설정
최소시청시간 10분에 맞추어 새롭게 T/F 컬럼을 만들어본다.


정우가 시청 기준에 맞는 T/F 컬럼을 줄테니 일단 코드를 짜기 위해 10분 기준으로 T/F 컬럼을 만들자

```python
# 컬럼 생성 1. 주차 구분: for 문 활용
won = []

for row in tidy_fin['프로그램시청시간']:
    if row >= 600:
        won.append(True)
    else:
        won.append(False)
```

```python
tidy_fin['WorN'] = won
```

```python
tidy_fin
```

# 개인 ID당 장르별 비율
##### 1. 우선 True 값만 다 모아보자

```python
t_user = tidy_fin[tidy_fin['WorN'] == True]
```

```python
t1 = t_user[t_user['기간'] == 'T1']
t2 = t_user[t_user['기간'] == 'T2']
t3 = t_user[t_user['기간'] == 'T3']
t4 = t_user[t_user['기간'] == 'T4']
```

##### 2. ID 별 장르 당 시청시간

```python
# 1기
t_genre_1 = t1.groupby(['기간', 'ID', '대분류'])['프로그램시청시간'].sum().reset_index(name='장르시청')
# 2기
t_genre_2 = t2.groupby(['기간', 'ID', '대분류'])['프로그램시청시간'].sum().reset_index(name='장르시청')
# 3기
t_genre_3 = t3.groupby(['기간', 'ID', '대분류'])['프로그램시청시간'].sum().reset_index(name='장르시청')
# 4기
t_genre_4 = t4.groupby(['기간', 'ID', '대분류'])['프로그램시청시간'].sum().reset_index(name='장르시청')
```

##### 3. 장르별 총합을 구해서 맵핑

```python
# 1기
t_sum_1 = t1.groupby(['ID'])['프로그램시청시간'].sum().reset_index(name='totv')
# 2기
t_sum_2 = t2.groupby(['ID'])['프로그램시청시간'].sum().reset_index(name='totv')
# 3기
t_sum_3 = t3.groupby(['ID'])['프로그램시청시간'].sum().reset_index(name='totv')
# 4기
t_sum_4 = t4.groupby(['ID'])['프로그램시청시간'].sum().reset_index(name='totv')
```

```python
# merge & prop
# 1기
t_genre_01 = t_genre_1.merge(t_sum_1)
t_genre_01['prop'] = t_genre_01['장르시청'] / t_genre_01['totv'] * 100
t_genre_01 = t_genre_01.drop(['장르시청', 'totv'], axis=1)
# 2기
t_genre_02 = t_genre_2.merge(t_sum_2)
t_genre_02['prop'] = t_genre_02['장르시청'] / t_genre_02['totv'] * 100
t_genre_02 = t_genre_02.drop(['장르시청', 'totv'], axis=1)
# 3기
t_genre_03 = t_genre_3.merge(t_sum_3)
t_genre_03['prop'] = t_genre_03['장르시청'] / t_genre_03['totv'] * 100
t_genre_03 = t_genre_03.drop(['장르시청', 'totv'], axis=1)
# 4기
t_genre_04 = t_genre_4.merge(t_sum_4)
t_genre_04['prop'] = t_genre_04['장르시청'] / t_genre_04['totv'] * 100
t_genre_04 = t_genre_04.drop(['장르시청', 'totv'], axis=1)
```

```python
t_genre_01
```

```python
# pivotting
# 1기
t_genre_01_pivot = t_genre_01.pivot_table(index=['기간', 'ID'], columns='대분류', values='prop')
t_genre_01_pivot = t_genre_01_pivot.reset_index()
# 2기
t_genre_02_pivot = t_genre_02.pivot_table(index=['기간', 'ID'], columns='대분류', values='prop')
t_genre_02_pivot = t_genre_02_pivot.reset_index()
# 3기
t_genre_03_pivot = t_genre_03.pivot_table(index=['기간', 'ID'], columns='대분류', values='prop')
t_genre_03_pivot = t_genre_03_pivot.reset_index()
# 4기
t_genre_04_pivot = t_genre_04.pivot_table(index=['기간', 'ID'], columns='대분류', values='prop')
t_genre_04_pivot = t_genre_04_pivot.reset_index()
```

##### 4. 이제 다 합쳐보자.

```python
genre_fin = pd.concat([t_genre_01_pivot, t_genre_02_pivot, t_genre_03_pivot, t_genre_04_pivot])
```

```python
genre_fin
```

# 개인 ID 당 채널 비율

#### 개인 ID당 채널 비율 => 상위 6개, 6개 비율의 총합은 100%가 아니다.
##### 1. 지상파 4사, 종편 2사인 걸 감안하고 만들어본다. 
##### 2. 유사하게 하되, 상위  채널이 무엇인지 알아본다. 

```python
# 전체 데이타프레임에서 상위 6개 채널 것만 쪼개서 사용해보자. 
kbs2 = tidy_fin[tidy_fin['방송사'] == ' KBS2']
jtbc = tidy_fin[tidy_fin['방송사'] == ' JTBC']
kbs1 = tidy_fin[tidy_fin['방송사'] == ' KBS1']
mbc = tidy_fin[tidy_fin['방송사'] == ' MBC']
sbs = tidy_fin[tidy_fin['방송사'] == ' SBS']
mbn = tidy_fin[tidy_fin['방송사'] == ' MBN']

frames = [kbs2, jtbc, kbs1, mbc, sbs, mbn]
top_ch = pd.concat(frames)
top_ch = top_ch[top_ch['WorN'] == True]
```

```python
# 여기서 다시 한 번 기간별로 나눠보자.
tc1 = top_ch[top_ch['기간'] == 'T1']
tc2 = top_ch[top_ch['기간'] == 'T2']
tc3 = top_ch[top_ch['기간'] == 'T3']
tc4 = top_ch[top_ch['기간'] == 'T4']
```

##### 3. 그 다음엔 ID와 채널을 기준으로 그룹바이 펑션을 사용해본다. 

```python
# 1기
t_channel_1 = tc1.groupby(['기간', 'ID', '방송사'])['프로그램시청시간'].sum().reset_index(name='채널시청')
# 2기
t_channel_2 = tc2.groupby(['기간', 'ID', '방송사'])['프로그램시청시간'].sum().reset_index(name='채널시청')
# 3기
t_channel_3 = tc3.groupby(['기간', 'ID', '방송사'])['프로그램시청시간'].sum().reset_index(name='채널시청')
# 4기
t_channel_4 = tc4.groupby(['기간', 'ID', '방송사'])['프로그램시청시간'].sum().reset_index(name='채널시청')
```

```python
t_channel_1
```

##### 4. ID의 기간별 시청 총합은 미리 구해놓은 것이 있으니 다음 단계로
merge를 진행하되, total 시청시간 데이터는 tidy_fin 중 시청했다고 볼 수 있는 것 전체로 (시청시간 > 600) 붙여준다.

```python
# 1기
top_c1 = t_channel_1.merge(t_sum_1)
top_c1['prop'] = top_c1['채널시청'] / top_c1['totv'] * 100
top_c1 = top_c1.drop(['채널시청', 'totv'], axis=1)

# 2기
top_c2 = t_channel_2.merge(t_sum_2)
top_c2['prop'] = top_c2['채널시청'] / top_c2['totv'] * 100
top_c2 = top_c2.drop(['채널시청', 'totv'], axis=1)

# 3기
top_c3 = t_channel_3.merge(t_sum_3)
top_c3['prop'] = top_c3['채널시청'] / top_c3['totv'] * 100
top_c3 = top_c3.drop(['채널시청', 'totv'], axis=1)

# 4기
top_c4 = t_channel_4.merge(t_sum_4)
top_c4['prop'] = top_c4['채널시청'] / top_c4['totv'] * 100
top_c4 = top_c4.drop(['채널시청', 'totv'], axis=1)
```

```python
top_c1
```

```python
# 이제 pivotting을 진행한다.
# 1기
top_c1_pivot = top_c1.pivot_table(index=['기간', 'ID'], columns='방송사', values='prop')
top_c1_pivot = top_c1_pivot.reset_index()
# 2기
top_c2_pivot = top_c2.pivot_table(index=['기간', 'ID'], columns='방송사', values='prop')
top_c2_pivot = top_c2_pivot.reset_index()
# 3기
top_c3_pivot = top_c3.pivot_table(index=['기간', 'ID'], columns='방송사', values='prop')
top_c3_pivot = top_c3_pivot.reset_index()
# 4기
top_c4_pivot = top_c4.pivot_table(index=['기간', 'ID'], columns='방송사', values='prop')
top_c4_pivot = top_c4_pivot.reset_index()
```

##### 5. 다 합친다

```python
channel_fin = pd.concat([top_c1_pivot, top_c2_pivot, top_c3_pivot, top_c4_pivot])
```

```python
channel_fin
```

```python
df_fin = genre_fin.merge(channel_fin)
```
